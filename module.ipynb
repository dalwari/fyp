```python
import easyocr
reader=easyocr.Reader(['en'])
```

```python
#import pytesseract as tess
from PIL import Image
from PIL import ImageDraw
#url='/content/ir1.jpg'
url='/content/res1.png'
#url='/content/dec.jpg'
#url='/content/hos3.png'
#url='/content/64img.png'
#url='/content/64img1.png'
#url='/content/hos.jpg'
img=Image.open(url)
img
```

```python
bounds=reader.readtext(url)
bounds 
```

```python
def draw_boxes(img,bounds,color='cyan',width=2):
  draw=ImageDraw.Draw(img)
  for bound in bounds:
    p0,p1,p2,p3=bound[0]
    draw.line([*p0,*p1,*p2,*p3,*p0],fill=color,width=width)
  return img
draw_boxes(img,bounds)
```
